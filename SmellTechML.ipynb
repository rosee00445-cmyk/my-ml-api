```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'enose_dataset_to_predict_human_disease.csv')
df.head()
```

```python
print(df.to_string())
```

```python
# In X_train, the first part (:) chooses all rows, and the second part (2:8)
# chooses columns excluding 8
# In Y_train, the first part (:) chooses all rows, and the second part (8)
# chooses the last column (target class)
# X_train = df.iloc[:,2:8]----
# Y_train = df.iloc[:,8]----

# X_train = df[['TGS2600','TGS2602','TGS2611','TGS2610','TGS2620','TGS826']]----
# y_train = df['Subjek']----

X_train = df[['TGS2600','TGS2602','TGS2611','TGS2610','TGS2620','TGS826']]
y_train = df['Subjek']

# X_test is used to construct a random row (or the patient's data)
X_test = [[23, 50, 10, 20, 30, 25]]
prediction = (X_test)
print("Prediction:", prediction)
```

```python
from sklearn.naive_bayes import GaussianNB
nb_gua = GaussianNB()
nb_gua.fit(X_train, y_train)
nb_gua.predict(X_test)
```
